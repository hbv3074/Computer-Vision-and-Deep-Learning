```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
path = '/kaggle/input/edge-detection/7d15b18b-1e0d6e3f.jpg'
image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

def plot(result, rtitle):
    """Utility function to plot the original and result images."""
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(image, cmap='gray')
    plt.subplot(1, 2, 2)
    plt.title(rtitle)
    plt.imshow(result, cmap='gray')
    plt.show()

while True:
    print("\nEdge Detection Options:")
    print("1. Median Filter")
    print("2. Box Filter")
    print("3. Gaussian Filter")
    print("4. Sobel Filter")
    print("5. Canny Edge Detection")
    print("6. Prewitt Operator")
    print("7. Exit")
    
    try:
        choice = int(input("Enter your choice: "))
    except ValueError:
        print("Invalid input. Please enter a number between 1 and 7.")
        continue

    if choice == 1:
        # Median Filter
        result = cv2.medianBlur(image, ksize=3)
        plot(result, "Median Filter")
    
    elif choice == 2:
        # Box Filter
        result = cv2.boxFilter(image, ddepth=-1, ksize=(3, 3))
        plot(result, "Box Filter")
    
    elif choice == 3:
        # Gaussian Filter
        sigma = 1  # You can take this as input if required
        result = cv2.GaussianBlur(image, (3, 3), sigma)
        plot(result, "Gaussian Filter")
    
    elif choice == 4:
        # Sobel Filter
        sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
        sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
        sobel_combined = cv2.magnitude(sobel_x, sobel_y)
        plot(sobel_combined, "Sobel Filter")
    
    elif choice == 5:
        # Canny Edge Detection
        result = cv2.Canny(image, threshold1=100, threshold2=200)
        plot(result, "Canny Edge Detection")
    
    elif choice == 6:
        # Prewitt Operator
        kernel_x = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
        kernel_y = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
        prewitt_x = cv2.filter2D(image, ddepth=-1, kernel=kernel_x)
        prewitt_y = cv2.filter2D(image, ddepth=-1, kernel=kernel_y)
        prewitt_combined = cv2.magnitude(prewitt_x.astype(float), prewitt_y.astype(float))
        plot(prewitt_combined, "Prewitt Operator")
    
    elif choice == 7:
        print("Exiting...")
        break
    
    else:
        print("Invalid choice. Please select a number between 1 and 7.")
```
